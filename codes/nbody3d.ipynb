```python
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load("lcdm_dm_structure_formation.npz")
```

```python
for key in data.keys():
  print(key)
```

```python
timezero = pd.DataFrame(data['dm_pos'][:,0,:])
timezero
```

```python
#ax = Axes3D(fig)
angleA = 10
angleB = 11
for index2 in range(0,100):
    timezero = pd.DataFrame(data['dm_pos'][:,index2,:])
    
    #1#
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111, projection='3d')
    ax.view_init(33, angleA)
    ax.set_axis_off()
    ax.set_zlim(-5000, 5000)
    ax.set_ylim(-5000, 5000)
    ax.set_xlim(-5000, 5000)
    for index in range(0, 50):
        ax.scatter(timezero.iloc[index][0], timezero.iloc[index][1], timezero.iloc[index][2], c='r', marker='^')

    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')

    #plt.show()
    plt.savefig("pics_sc1\pic" + str(index2) + ".png")
    plt.close('all')
    
    #2#
    
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111, projection='3d')
    ax.view_init(33, angleB)
    ax.set_axis_off()
    ax.set_zlim(-5000, 5000)
    ax.set_ylim(-5000, 5000)
    ax.set_xlim(-5000, 5000)
    for index in range(0, 50):
        ax.scatter(timezero.iloc[index][0], timezero.iloc[index][1], timezero.iloc[index][2], c='r', marker='^')

    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_zlabel('Z Label')

    #plt.show()
    plt.savefig("pics_sc2\pic" + str(index2) + ".png")
    plt.close('all')
```
